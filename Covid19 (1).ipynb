```python
#Importing libraries
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_death_rates_by_country"  #Website url
```

```python
#Checking page availability
page = requests.get(url)
print(page)
```

```python
#Getting website codes
soup = BeautifulSoup(page.text,"html")
soup
```

```python
#finding all tables in the site
tables = soup.find_all("table")
tables
```

```python
#Extracting the specific table
table = soup.find_all("table")[1]
table
```

```python
#Extracting table headers (Collumn names)
table_column = table.find_all("tr")[0]
table_column
```

```python
#Converting column names into list
column_names = [tittle.text.strip() for tittle in table_column]
column_names
```

```python
#Extracting required columns
selected_columns = [column_names[i] for i in [1,3,5,7]]
selected_columns
```

```python
#Creating dataframe
import pandas as pd
```

```python
Data = pd.DataFrame(columns = selected_columns)
```

```python
#Viewing dataFrame
```

```python
Data
```

```python
#Extracting entries oF each column in table
```

```python
column_data = table.find_all("tr")[1:]
column_data
```

```python
#Extracting table headers and table data from the rows as list 
for row in column_data[:-1]:                                 #Last list is removed 
    row_data = row.find_all(["th" , "td"])
    individual_row_data = [row.text.strip() for row in row_data]
    print(individual_row_data)
```

```python
#Extracting table data ::Deaths/million,Deaths,Cases
for row1 in column_data:
    row_data1 = row1.find_all("td")
    individual_row_data1 = [row1.text.strip() for row1 in row_data1]
    print(individual_row_data1)
```

```python
#Extracting table data headers ::Country
for row in column_data[:-1]:
    row_data = row.find_all("th")
    individual_row_data = [row.text.strip() for row in row_data]
    print(individual_row_data)
```

```python
#Extracting data of each row th:Country , td ::Deaths/million,Deaths, Cases then inserting in Dataframe (Data)
for row in column_data[:-1]:
    row_data = row.find_all(["th", "td"])
    individual_row_data = [row.text.strip() for row in row_data]
    Length = len(Data)
    Data.loc[Length] = individual_row_data
```
