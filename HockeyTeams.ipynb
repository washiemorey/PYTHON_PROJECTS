# Extracting Hockey Teams tables on pages 1,2,3,4 and 5 and storing in dataframe then on my PC as csv file

```python
#Calling libraries used in scripping website
from bs4 import BeautifulSoup
import requests
```

```python
#Website url   Page1
url = "https://www.scrapethissite.com/pages/forms/?page_num=1"
```

```python
#Checking page availability
page = requests.get(url)
page
```

```python
#Extracting website codes
soup = BeautifulSoup(page.text,"html")
soup
```

```python
#Checking tables in the website
tables = soup.find_all("table")
tables
```

```python
#Extracting specific tables 
table = soup.find_all("table")[0]
table
```

# Extracting table column names 

```python

table_columns = table.find_all("tr")[0]
table_columns
```

```python

column_names = table_columns.find_all("th")
list_column_names = [tittle.text.strip() for tittle in column_names]
list_column_names
```

# Creating data frame with name Data

```python
#Creating data frame 
import pandas as pd
Data = pd.DataFrame(columns = list_column_names)
Data
```

# Extracting data oF each columns from table on page 1  :: table data rows

```python

table_columns = table.find_all("tr")[1:]
table_columns
```

```python
for row in table_columns:
    row_data = row.find_all("td")
    individual_row_data = [row.text.strip() for row in row_data]
    print(individual_row_data)
```

```python
#Adding row in the pandas dataframe Data
for row in table_columns:
    row_data = row.find_all("td")
    individual_row_data = [row.text.strip() for row in row_data]
    Length = len(Data)
    Data.loc[Length] = individual_row_data
```

```python
#Viewing Dataframe
Data
```
